```python
from PIL import Image
import numpy as np

# Load the grayscale image
image_path = "Test.tif"  #actual path to image
gray_image = Image.open(image_path).convert("L")

# Convert the image to a NumPy array
image_array = np.array(gray_image)

# Display the shape of the array (height x width)
print("Image Shape:", image_array.shape)

# Display the 2D array
print("Image Array:")
print(image_array)
```

```python
# Define the size of the squares
square_size = 200

# Get the dimensions of the image
height, width = image_array.shape

# Create an empty list to store the 200x200 pixel squares
squares = []

# Loop through the image and extract 200x200 pixel squares
for i in range(0, height, square_size):
    for j in range(0, width, square_size):
        square = image_array[i:i+square_size, j:j+square_size]
        squares.append(square)

```

```python
# Create an empty list to store the calculated values for each square
square_stats = []

# Loop through the image and calculate statistics for each 200x200 pixel square
for i in range(0, height, square_size):
    for j in range(0, width, square_size):
        square = image_array[i:i+square_size, j:j+square_size]

        # Calculate mean intensity, standard deviation, mean luminosity, and mean contrast
        mean_intensity = np.mean(square)
        std_intensity = np.std(square)
        mean_luminosity = (np.min(square) + np.max(square)) / 2
        mean_contrast = (np.max(square) - np.min(square)) / 2

        # Append the statistics to the list
        square_stats.append({
            "Mean Intensity": mean_intensity,
            "Std Intensity": std_intensity,
            "Mean Luminosity": mean_luminosity,
            "Mean Contrast": mean_contrast
        })

# Print the statistics for each square
for i, stats in enumerate(square_stats):
    print(f"Stats for Square {i + 1}:")
    for key, value in stats.items():
        print(f"{key}: {value}")
    print()
```

```python
from PIL import Image
import numpy as np

# Load the grayscale image
image_path = "Test.tif"  # Replace with the actual path to your image
gray_image = Image.open(image_path).convert("L")
image_array = np.array(gray_image)

# Define the size of the squares
square_size = 200

# Get the dimensions of the image
height, width = image_array.shape

# Create an empty array to store the binary classification for each pixel
binary_classification = np.zeros_like(image_array, dtype=np.uint8)

# Threshold value
threshold = 1

# Loop through the image and classify pixels into 'background_pixel' and 'foreground_pixel'
for i in range(0, height, square_size):
    for j in range(0, width, square_size):
        square = image_array[i:i+square_size, j:j+square_size]
        square_distances = np.abs((square - np.mean(square)) / np.std(square))
        
        # Apply thresholding
        binary_mask = square_distances > threshold
        
        # Assign categories based on the threshold
        binary_classification[i:i+square_size, j:j+square_size] = np.where(binary_mask, 255, 0)

# Apply the intensity transformation for pixels in the 'background_pixel' category
for i in range(height):
    for j in range(width):
        if binary_classification[i, j] == 0:  # 'background_pixel'
            original_intensity = image_array[i, j]
            mean_luminosity = np.mean(image_array[i//square_size * square_size:(i//square_size + 1) * square_size,
                                                 j//square_size * square_size:(j//square_size + 1) * square_size])
            mean_contrast = (np.max(square) - np.min(square)) / 2
            new_intensity = (original_intensity - mean_luminosity) / mean_contrast
            image_array[i, j] = new_intensity

# Now, 'image_array' contains the modified intensities for pixels in the 'background_pixel' category

# Print the modified intensity values for the first 100 pixels
print("Modified Intensity Values for the first 100 pixels:")
print(image_array[:10, :10])  # Adjust the slice to show the desired region
```
